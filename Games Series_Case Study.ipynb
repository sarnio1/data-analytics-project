#Games Series Data

##Preparing the Packages

```python
#import package and its alias

import pandas as pd
```

##Reading and Cleaning the Data

```python
#import data from Google Sheets

sheet_url = 'https://docs.google.com/spreadsheets/d/1GQmkKv3E3k6_OyTuCxBpd7thWwpve7HJed63Jp_Id3A/edit#gid=1485085913'
sheet_url_trf = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
print(sheet_url_trf)

df = pd.read_csv(sheet_url_trf)
df.head(10)
```

```python
df.info()
```

The data type of 'Release' is incorrect. It is formatted as an object instead of datetime, so we will change it.

```python
#change data type

df_cleaned = df.copy()
df_cleaned['Release'] = pd.to_datetime(df_cleaned['Release'])
print(df_cleaned.info())
df_cleaned.head()
```

Now, the 'Release' data has been formatted as datetime data type.

```python
#check typo by counting the amount of variety of the data


for col_name in ['Genre', 'Developer', 'Publisher']:
        print(df_cleaned[col_name].value_counts(), '\n')
```

The data is large in variety.

```python
#check duplicated data

df_cleaned['Name'].duplicated()
```

There are 2 rows where duplicated data exist, we are going to remove these duplicated data while keeping the original ones.

```python
#remove duplicated data

df_removed = df_cleaned.drop_duplicates(subset=['Name'], keep='first')
df_removed.tail()
```

Now, the data has been cleaned. We have changed the type of each data properly (changing the string data in column 'Release' with datetime format) and managed to identify duplicated data in column 'Name'.The reason we tested only the 'Name' column is that it contains unique values that is not repeated in the list.

##Exploratory Data Analysis

###Which game is the oldest and the newest in the dataset?

```python
df_removed.sort_values(by='Release').head(3)
```

The oldest game in the dataset is Hydlide, as it was released in 1984.

```python
df_removed.sort_values(by='Release', ascending=False).head(3)
```

Whereas the newest game in the dataset is Valheim. It was released in 2021.

###Which publisher released most of the game?

```python
print(df_removed['Publisher'].value_counts())
```

The publisher that released the most games in the dataset is Electronic Arts, with a total of 19 games.

###Which developer developed most of the games?

```python
print(df_removed['Developer'].value_counts())
```

The developer that developed the most games in the dataset is Blizzard Entertainment, with a total of 8 games.

###Which series has the most sales?

```python
agg_sales = df_removed.groupby('Series', as_index=False).agg({'Sales':'sum'}).sort_values(by='Sales', ascending=False)
agg_sales.head()
```

The series that have the most sales in the dataset is Minecraft, with a total sales of 33.0.

###Which series have the most games?

```python
agg_games = df_removed.groupby('Series', as_index=False).agg({'Name':'nunique'}).sort_values(by='Name', ascending=False)
agg_games.head()
```

The series that have the most games in the dataset is Command & Conquer, with a total games of 5.
